# Question Answers

difference in speed for the 50km, 50mi male to female

```python
df.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

what age froup are the best in the 50m race (20 + races min)(show 15)

```python
df.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(15)
```

what age froup are the worst in the 50m race (20 + races min)(show 20)

```python
df.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count>19').head(20)
```

# Seasons for the data -> Slower in summer than winter?
spring 3-5 \
summer 6-8 \
fall 9-11 \
winter 12-2 

```python
df['race_month'] = df['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df.head(25)
```

```python
df['race_season'] = df['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter') 
```

```python
df.head(25)
```

```python
df.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

50mi only

```python
df.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
